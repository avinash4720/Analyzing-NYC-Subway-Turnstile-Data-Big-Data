```python
from urllib import request
import os
import pandas as pd
```

```python
monthsNonLeapYear = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
monthsLeapYear = {1:31, 2:29, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
```

```python
initialList = ['141018', '141025', '141101', '141108', '141115', '141122', '141129', '141206', '141213', '141220', '141227']
for dates in initialList:
    remote_url = f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_" + f"{dates}" + f".txt"
    local_file = f"/Users/arjun17/Desktop/BigData Datasets/" + f"{dates[4:6]}" + "_" + f"{dates[2:4]}" + "_" + f"{dates[0:2]}" + f".txt"
    request.urlretrieve(remote_url, local_file)
start = 3
exit = 0
for i in range(15, 24):
    if i % 4 == 0:
        for k in range(1, 13):
            while start <= monthsLeapYear[k]:
                remote_url = f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_" + f"{i:02d}" + f"{k:02d}" + f"{start:02d}" + f".txt"
                local_file = f"/Users/arjun17/Desktop/BigData Datasets/" + f"{start:02d}" + "_" + f"{k:02d}" + "_" + f"{i:02d}" + f".txt"
                request.urlretrieve(remote_url, local_file)
                start += 7
            start %= monthsLeapYear[k]
    else:
        for k in range(1, 13):
            while start <= monthsNonLeapYear[k]:
                remote_url = f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_" + f"{i:02d}" + f"{k:02d}" + f"{start:02d}" + f".txt"
                local_file = f"/Users/arjun17/Desktop/BigData Datasets/" + f"{start:02d}" + "_" + f"{k:02d}" + "_" + f"{i:02d}" + f".txt"
                request.urlretrieve(remote_url, local_file)
                if f"{i:02d}" + f"{k:02d}" + f"{start:02d}" == "230429":
                    exit = 1
                    break
                start += 7
            if exit == 1:
                break
            start %= monthsNonLeapYear[k]
        if exit == 1:
            break
        
```

```python
dir_list = os.listdir("/Users/arjun17/Desktop/BigData Datasets/")
```

```python
years = {2014:[], 2015:[], 2016:[], 2017:[], 2018:[], 2019:[], 2020:[], 2021:[], 2022:[], 2023:[]}
for j in range(14, 24):
    for val in dir_list:
        if str(j) == val[6:8]:
            years[int("20" + val[6:8])].append(val)
            continue
```

```python
for k in years:
    df_csv_append = pd.DataFrame()
    for j in years[k]:
        df = pd.read_csv("/Users/arjun17/Desktop/BigData Datasets/" + j)
        df_csv_append = df_csv_append.append(df, ignore_index=True)
    df_csv_append.to_csv("/Users/arjun17/Desktop/BigData Datasets/20"+j[6:8]+".csv")  
```
